## Phase Diagrams for One Dimensional Nonlinear Differential Equations 

So, as discussed in class, for nonlinear problems of the form 

$$
\frac{dy}{dt} = f(y)
$$

a very effective way to study these problems is through studying and classifying their equilibrium solutions, i.e. those constant solutions $y(t) = y_{e}$ such that $f(y_{e})=0$.  

If we look at how the sign of $f(y)$ changes around each equilibrium point, we can assess its _stability_, which is to say, are solutions attracted to or repelled away from a given equilibrium point.  

Below, we look at an example in which $f(y)=\cos(4y)$, $0\leq y \leq 2\pi$.

```python
fun = lambda y: np.cos(4.*y)
a = 0.
b = 2.*np.pi
myplotter = PP.PhasePlotter(a, b, fun)
```

```python
myplotter.PhaseField1D()
```

As can be seen from this example, there are in the given interval a total of eight equilibrium points which alternate with respect to their stability with the stable/attractive points given by solid dots and the unstable/repellent points given by clear dots.  The arrows on the horizontal axis reflect the sign of $f(y)$.  When $f(y)$ is blue, it is greater than zero, and therefore the arrows point to the right.  When $f(y)$ is red, it is less than zero, and therefore the arrows point to the left.
